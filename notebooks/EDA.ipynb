```python

#Importing the necessary libraries
import numpy as np
import streamlit as st
import pandas as pd
import plotly.express as px
import pyarrow as pa
import matplotlib.pyplot as plt

#Loading the data
df = pd.read_csv('C:/Users/hamed/OneDrive/Documents/Data - TripleTen/VS/Triple_Ten_sprint4_project/vehicles_us.csv')
# Checking some information about the data
df.describe()
```

```python
df.info()
```

```python
#Taking a look at the 10 Sample rows of the data
df.sample(10)
```

```python
#Checking the duplicates in the data
df.duplicated().sum()
```

```python
# Filling the missing values in "is_4wd", "paint_color", and "odometer" columns
df['is_4wd'] = df['is_4wd'].fillna(0)
df['paint_color'] = df['paint_color'].fillna('unknown')

# Filling the odometer, cylinders and model_year column with the mean of the column
df['odometer'] = df['odometer'].fillna(df['odometer'].mean())
df['cylinders'] = df['cylinders'].fillna(df['cylinders'].mean())    
df['model_year'] = df['model_year'].fillna(df['model_year'].mean())


# Checking the missing values in the columns again to make sure that the missing values are filled
print(f"The qty of null cell on is_4wd column: {df['is_4wd'].isnull().sum()}")
print(f"The qty of null cell on paint_color column: {df['paint_color'].isnull().sum()}")
print(f"The qty of null cell on odometer column: {df['odometer'].isnull().sum()}")
print(f"The qty of null cell on cylinders column: {df['cylinders'].isnull().sum()}")
print(f"The qty of null cell on model_year column: {df['model_year'].isnull().sum()}")
```

```python
# Double check the data
df.info()
```

```python
# Using the model column to create a new column called manufacturer that has only name of the brand.
df['manufacturer'] = df['model'].apply(lambda x: x.split()[0])  # Splitting the model column and taking the first word as the manufacturer
df
```

```python
#CHECKING THE UNIQUE MANUFACTURERS Values
print(df['manufacturer'].unique())
```

Conducting an exploration to uncover patterns and trends.

```python
# Creating a plot that shows the number of cars per brand in this dataset
fig, ax = plt.subplots(figsize=(10, 8))

x = df['manufacturer'].value_counts().index # X values of the plot
ax.bar(x, df['manufacturer'].value_counts()) # Creating the bar plot

ax.set_title("Number of Cars Per Brand") # Title of the plot
ax.set_xlabel("Manufacturer") # X label of the plot
ax.set_ylabel("Number of cars") # Y label of the plot
plt.xticks(rotation=45) # Rotating the x labels to make them more readable
plt.savefig('no_car_per_brand.png')    # Save the plot as a png file as It is not antractive in the streamlit. Please check the file in the same folder
```

The above chart is a bar graph titled "Number of Cars Per Brand", displaying the distribution of cars across various manufacturers. 
Here's a summary:

**Top Manufacturers:**<br>
Ford is the leading manufacturer, with the highest number of cars, exceeding 12,000 units.
Chevrolet and Toyota follow as the second and third highest, respectively, though both have noticeably fewer cars than Ford.

**Mid-Tier Manufacturers:**<br>

Brands like Honda, Ram, Jeep, and Nissan form the mid-range group, each with a relatively comparable number of cars (between 3,000–5,000 units).
GMC and Subaru show slightly lower counts in this category.

**Lower-End Manufacturers:**<br>
Brands like Hyundai, Volkswagen, Chrysler, and Kia have fewer cars, typically under 2,000 units each.<br>
Premium brands like Cadillac, BMW, Acura, and Mercedes-Benz have the smallest presence, with significantly fewer cars.

**Trends:**<br>
There is a steep decline in the number of cars as you move from the most popular brands (Ford) to the least represented brands (Mercedes-Benz).
This distribution reflects that a few dominant manufacturers hold the majority of the market share.

```python
# Creating a plot that shows the average price of cars per brand in this dataset
fig, ax = plt.subplots(figsize=(10, 8))
x = df['manufacturer'].value_counts().index # X values of the plot
ax.bar(x, df.groupby('manufacturer')['price'].mean()) # Creating the bar plot

ax.set_title("Average Price of Cars Per Brand") # Setting the title of the plot
ax.set_xlabel("Manufacturer") # Setting the x-axis label
ax.set_ylabel("Average Price") # Setting the y-axis label
plt.xticks(rotation=90) # Rotating the x-axis labels to make them readable
plt.savefig('avg_price_per_brand.png')    # Save the plot as a png file as It is not antractive in the streamlit. Please check the file in the same folder
```

This chart is titled "Average Price of Cars Per Brand" and presents the average price of cars for various manufacturers. 
**Highest Average Price:**<br>
Kia has the highest average price, reaching approximately $35,000, making it a clear outlier in this dataset.

**Premium Brands:**<br>
Cadillac and BMW also show high average prices, both exceeding $20,000, indicating their premium positioning in the market.

**Mid-Range Prices:**
Brands such as Honda, Ram, GMC, and Subaru have average prices between $15,000–$20,000, catering to mid-market consumers.
Lower Average Prices:

Ford, Chevrolet, and Toyota are in the lower price range, averaging between $10,000–$15,000, aligning with their broader market focus on affordability.

**Economy Segment:**<br>
Brands like Hyundai, Dodge, Volkswagen, and Jeep fall in the lower price range, likely offering economical models.
Insights:

Mercedes-Benz, typically considered a luxury brand, has a surprisingly low average price in this dataset, similar to brands like Acura and Nissan.
The chart shows significant price variation among manufacturers, with Kia being the standout outlier.

```python
# Creating a plot that shows the color of cars per brand in this dataset
fig, ax = plt.subplots(figsize=(10, 8))
ax.hist(df['paint_color'], bins=len(df['paint_color'].unique())) # Creating the histogram plot
ax.set_title("Number of Cars Per Paint Color") # Setting the title of the plot
ax.set_xlabel("Paint Color") #  Setting the x-axis label
ax.set_ylabel("Number of cars") # Setting the y-axis label
plt.xticks(rotation=45) #   Rotating the x-axis labels to make them readable
plt.savefig('color_per_brand.png')   # Save the plot as a png file as It is not attractive in the streamlit. Please check the file in the same folder
```

The histogram provides a visual representation of the distribution of cars based on their paint colors. Here's a summary of the data presented:

**Dominant categories:**<br>
"White" paint colors have the highest counts, each exceeding 9,000 cars.

"Black" and "Silver" also have significant counts, though lower than "White"."

**Moderate categories:**<br>
"Red," "Blue," and "Gray" fall in the middle range, with counts around 4,000–6,000 cars.

**Less common categories:**<br>
"Brown," "Green," and "Yellow" paint colors have noticeably fewer cars.
"Custom," "Orange," and "Purple" are the least represented categories.

**General trend:**<br>
The majority of cars fall into a few popular categories (e.g., "White," "Black," and "Silver").
Niche colors like "Purple" and "Custom" are rare.
